```python
import pandas as pd
df = pd.read_csv("FinalData.csv", sep=";")
```

```python
for method in ["Centroid_E5", "CLS_E5", "VLDDR_E5"]:
    same = df.loc[df["same"] == 1, method]
    different = df.loc[df["same"] == 0, method]

    max_different = different.max()
    print(f"Amount of sentence pairs above threshold for {method}: {(same > max_different).sum()}")
```
