```python
import pandas as pd
df = pd.read_csv("FinalData.csv", sep=";").fillna(0)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu
import numpy as np

plt.rcParams.update({
    "font.size": 10,
    "axes.titlesize": 14,
    "axes.labelsize": 12,
    "xtick.labelsize": 11,
    "ytick.labelsize": 11,
    "legend.fontsize": 12,
})

fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# Left: non-contextual distances

# Calculate the significance
U, p = mannwhitneyu(df["non_contextual_dist_E5"], df["non_contextual_dist_GPT"])
E5_median = np.median(df["non_contextual_dist_E5"])
GPT_median = np.median(df["non_contextual_dist_GPT"])
print("For raw embeddings:")
print(f"Median E5: {E5_median:.2f}, Median GPT-2: {GPT_median:.2f}, U = {U}, p = {p:.4f}")

axes[0].hist(df["non_contextual_dist_E5"], label="E5", bins=30)
axes[0].hist(df["non_contextual_dist_GPT"], label="GPT-2", bins=30, alpha=0.9)
axes[0].set_title("Distribution of Raw Embedding Distances")
axes[0].set_xlabel("Distance Scores", fontweight="bold")
axes[0].set_ylabel("Count", fontweight="bold")
axes[0].legend()

# Calculate the significance
U, p = mannwhitneyu(df["contextual_dist_E5"], df["contextual_dist_GPT"])
E5_median = np.median(df["contextual_dist_E5"])
GPT_median = np.median(df["contextual_dist_GPT"])
print("For contextual embeddings")
print(f"Median E5: {E5_median:.2f}, Median GPT-2: {GPT_median:.2f}, U = {U}, p = {p:.4f}")

# Right: contextual distances
axes[1].hist(df["contextual_dist_E5"], label="E5", bins=30)
axes[1].hist(df["contextual_dist_GPT"], label="GPT-2", bins=30, alpha=0.9)
axes[1].set_title("Distribution of Contextual Distances")
axes[1].set_xlabel("Distance Scores", fontweight="bold")
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 5), sharey=True)

# Centroid
same = df.loc[df["same"] == 1, "Centroid_GPT"]
not_same = df.loc[df["same"] == 0, "Centroid_GPT"]

xmin = min(same.min(), not_same.min())
xmax = max(same.max(), not_same.max())
bins = np.linspace(xmin, xmax, 30)

axes[0].hist(not_same, label="Dissimilar sentences", bins=bins, alpha=1)
axes[0].hist(same, label="Similar sentences", bins=bins, alpha=0.8)
axes[0].set_title("Distribution of Centroid values")
axes[0].set_ylabel("Count", fontweight='bold')
axes[0].set_xlabel("Similarity Score", fontweight='bold')
axes[0].legend()

# Calculate the significance
U, p = mannwhitneyu(same, not_same)
same_median = np.median(same)
not_same_median = np.median(not_same)

# Calculate effect size
n1 = len(same)
n2 = len(not_same)
N = n1 + n2
mean_u = (n1 * n2) / 2
std_u = np.sqrt((n1 * n2 * (N + 1)) / 12)
z_score = abs((U - mean_u) / std_u)
r = z_score / np.sqrt(N)

print("For Centroid:")
print(f"$\\text{{Mdn}}_{{\\text{{Low-Similarity}}}} = {not_same_median:.4f}, \\text{{Mdn}}_{{\\text{{High-Similarity}}}} = {same_median:.4f}, U = {U}, p = {p:.4f}, r = {r:.2f}$")
print(f"Median High Similarity: {same_median:.4f}, Median Low Similarity: {not_same_median:.4f}, U = {U}, p = {p:.4f}")

# EOS
same = df.loc[df["same"] == 1, "EOS_GPT"]
not_same = df.loc[df["same"] == 0, "EOS_GPT"]

xmin = min(same.min(), not_same.min())
xmax = max(same.max(), not_same.max())
bins = np.linspace(xmin, xmax, 30)

axes[1].hist(not_same, label="Dissimilar sentences", bins=bins, alpha=1)
axes[1].hist(same, label="Similar sentences", bins=bins, alpha=0.8)
axes[1].set_title("Distribution of EOS values")
axes[1].set_xlabel("Similarity Score", fontweight='bold')
axes[1].legend()

# Calculate the significance
U, p = mannwhitneyu(same, not_same)
same_median = np.median(same)
not_same_median = np.median(not_same)

# Calculate effect size
n1 = len(same)
n2 = len(not_same)
N = n1 + n2
mean_u = (n1 * n2) / 2
std_u = np.sqrt((n1 * n2 * (N + 1)) / 12)
z_score = abs((U - mean_u) / std_u)
r = z_score / np.sqrt(N)

print("For EOS:")
print(f"$\\text{{Mdn}}_{{\\text{{Low-Similarity}}}} = {not_same_median:.4f}, \\text{{Mdn}}_{{\\text{{High-Similarity}}}} = {same_median:.4f}, U = {U}, p = {p:.4f}, r = {r:.2f}$")
print(f"Median High Similarity: {same_median:.4f}, Median Low Similarity = {not_same_median:.4f}, U = {U}, p = {p:.4f}")

# VLDDR

same = df.loc[df["same"] == 1, "VLDDR_GPT"]
not_same = df.loc[df["same"] == 0, "VLDDR_GPT"]

xmin = min(same.min(), not_same.min())
xmax = max(same.max(), not_same.max())
bins = np.linspace(xmin, xmax, 30)

axes[2].hist(not_same, label="Dissimilar sentences", bins=bins, alpha=1)
axes[2].hist(same, label="Similar sentences", bins=bins, alpha=0.8)
axes[2].set_title("Distribution of VLDDR values")
axes[2].set_xlabel("VLDDR Score", fontweight='bold')
axes[2].legend()

# Calculate the significance
U, p = mannwhitneyu(same, not_same)
same_median = np.median(same)
not_same_median = np.median(not_same)

# Calculate effect size
n1 = len(same)
n2 = len(not_same)
N = n1 + n2
mean_u = (n1 * n2) / 2
std_u = np.sqrt((n1 * n2 * (N + 1)) / 12)
z_score = abs((U - mean_u) / std_u)
r = z_score / np.sqrt(N)

print("For VLDDR:")
print(f"$\\text{{Mdn}}_{{\\text{{Low-Similarity}}}} = {not_same_median:.2f}, \\text{{Mdn}}_{{\\text{{High-Similarity}}}} = {same_median:.2f}, U = {U}, p = {p:.4f}, r = {r:.2f}$")
print(f"Median High Similarity: {same_median:.2f}, Median Low Similarity: {not_same_median:.2f}, U = {U}, p = {p:.4f}")

plt.tight_layout()
plt.show()
```
