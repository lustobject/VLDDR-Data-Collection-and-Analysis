```python
import pandas as pd
df = pd.read_csv("FinalData.csv", sep=";")

df_same = df[df["same"] == 0]
df_diff = df[df["same"] == 1]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = (18, 5)

# Increase default font sizes
plt.rcParams["font.size"] = 18
plt.rcParams["axes.titlesize"] = 22
plt.rcParams["axes.labelsize"] = 18
plt.rcParams["xtick.labelsize"] = 16
plt.rcParams["ytick.labelsize"] = 16
plt.rcParams["legend.fontsize"] = 16

y_columns = ["non_contextual_dist_GPT", "contextual_dist_GPT", "VLDDR_GPT"]
titles = ["Raw Embeddings", "Contextual Embeddings", "VLDDR"]
y_labels = ["Chordal Distance", "Chordal Distance", "VLDDR Values"]

for i, col in enumerate(y_columns, 1):
    plt.subplot(1, 3, i)
    
    sns.regplot(
        x=df["similarity_score"], 
        y=df[col], 
        line_kws={"color": "red", "label": "Linear Fit"}
    )
    
    plt.title(titles[i - 1], fontsize=22, fontweight='bold')
    plt.xlabel("Human Annotated Similarity Score", fontweight='bold', fontsize=18)
    plt.ylabel(y_labels[i - 1], fontweight='bold', fontsize=18)
    plt.legend(fontsize=16)

plt.tight_layout()
plt.show()
```
